```python
# 상길북 65번(리트코드 704번)

class Solution:
    def search(self, nums: List[int], target: int) -> int:
        # 왼쪽 끝과 오른쪽 끝 설정
        left, right = 0, len(nums) - 1
        # 중간값 설정
        while left <= right:
            mid = (left + right) // 2
            
            # nums 리스트의 중간 순서 값이 목표값보다 작을 때 왼쪽 값은 중간값에 1을 더한 값
            if nums[mid] < target:
                left = mid + 1
            # nums 리스트의 중간 순서 값이 목표값보다 작을 때 오른쪽 값은 중간값에 1을 뺀 값
            elif nums[mid] > target:
                right = mid - 1
            # 타겟이 목표값과 동일한 경우 중간값 반환
            else:
                return mid
        # 위의 조건이 모두 안 들어 맞으면 -1 반환
        return -1
```

```python
# 상길북 66번(리트코드 33번)

class Solution:
    def search(self, nums: List[int], target: int) -> int:
        # 예외 처리
        if not nums:
            return -1
        
        # 최솟값을 찾아 피벗 설정
        # 왼쪽 값 및 오른쪽 값 양쪽 끝 값으로 설정
        left, right = 0, len(nums) - 1
        # 중간값 설정
        while left < right:
            mid = left + (right - left) // 2
            
            # nums 리스트의 중간 순서 값이 오른쪽 순서 값보다 클 때 왼쪽 값 중간값+1로 변경
            if nums[mid] > nums[right]:
                left = mid + 1
            # nums 리스트의 중간 순서 값이 오른쪽 순서 값과 같거나 작을 때 오른쪽 값 중간값-1로 변경
            else:
                right = mid
        # while문 통과 후 피벗값 왼쪽 값으로 설정        
        pivot = left
        
        # 피벗 기준 이진 검색
        # 왼쪽 값 및 오른쪽 값 초기화
        left, right = 0, len(nums) - 1
        # 왼쪽값이 오른쪽 값과 같거나 작을 때 중간값은 left + (right - left)//2와 같음
        # 중간 피벗값은 (mid + pivot)을 nums 리스트의 길이로 나눈 나머지와 같음
        while left <= right:
            mid = left + (right - left) // 2
            mid_pivot = (mid + pivot) % len(nums)
            
            # nums 리스트의 중간 피벗값 순서의 값이 타겟보다 작을 경우 왼쪽값 중간값에 1을 더한 값으로 변경
            if nums[mid_pivot] < target:
                left = mid + 1
            # nums 리스트의 중간 피벗값 순서의 값이 타겟보다 클 경우 오른쪽값 중간값에 1을 뺀 값으로 변경    
            elif nums[mid_pivot] > target:
                right = mid - 1
            # nums 리스트의 중간 피벗값 순서의 값이 타겟과 같을 경우 중간 피벗값 반환
            else:
                return mid_pivot
        
        # 위의 모든 조건 통과 시 -1 반환
        return -1
```

```python
# 상길북 67번(리트코드 349번)

class Solution:
    def intersection(self, nums1: List[int], nums2: List[int]) -> List[int]:
        result: Set = set()
        # 양쪽 모두 정렬
        nums1.sort()
        nums2.sort()
        # 투 포인터 초깃값 설정
        i = j = 0
        # 투 포인터 nums1 및 nums2 범위 내에서 우측으로 이동하며 일치 여부 판별
        while i < len(nums1) and j < len(nums2):
            # nums1의 i번째 값이 nums2의 j번째 값보다 클 때 j 값 1 증가
            if nums1[i] > nums2[j]:
                j += 1
            # nums1의 i번째 값이 nums2의 j번째 값보다 작을 때 i 값 1 증가
            elif nums1[i] < nums2[j]:
                i += 1
            # 둘이 같을 때 result에 nums1의 i번째 값 추가 후 i, j 각각 1 증가
            else:
                result.add(nums1[i])
                i += 1
                j += 1
        
        # 모든 조건 통과 후 결과값 반환
        return result
```

```python
# 상길북 69번(리트코드 240번)

class Solution:
    def searchMatrix(self, matrix, target):
        # 예외 처리
        if not matrix:
            return False
        
        # 첫 행의 맨 뒤
        row = 0
        col = len(matrix[0]) - 1
        
        # row가 (matrix의 길이)-1과 같거나 작고 col 0과 같거나 클 때
        while row <= len(matrix) - 1 and col >= 0:
            # 타겟이 매트릭스 값과 같을 때 True 반환
            if target == matrix[row][col]:
                return True
            # 타겟이 작으면 왼쪽으로 이동
            elif target < matrix[row][col]:
                col -= 1
            # 타겟이 크면 아래로 이동
            elif target > matrix[row][col]:
                row += 1
        
        # 모든 조건 통과 시 False 반환
        return False
```
